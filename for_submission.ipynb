```python
import joblib
model_pretrained = joblib.load("Titanic-20240128.pkl")

import pandas as pd
df_test = o=pd.read_csv('test.csv')
```

```python
df_test.head()
```

```python
df_test.drop(["Name", "Ticket"], axis=1,  inplace=True)
df_test.drop("Cabin", axis=1, inplace=True)
df_test['Age'].fillna(df_test.groupby('Sex')['Age'].transform('median'), inplace=True)
```

```python
 df_test.isnull().sum()
```

```python
df_test.fillna(df_test["Fare"].mode()[0], inplace=True)
df_test.isnull().sum()
```

```python
df_test = pd.get_dummies(data=df_test, columns=['Sex', 'Embarked'])
df_test
```

```python
df_test.drop(["Sex_female", "Pclass"], axis=1, inplace=True)
```

```python
prediction2 = model_pretrained.predict(df_test)
```

```python
# prepare for submission
forSubmissionDF = pd.DataFrame(columns=["PassengerId", "Survived"])
forSubmissionDF["PassengerId"] = df_test["PassengerId"]
forSubmissionDF["Survived"] = prediction2
```

```python
forSubmissionDF.to_csv("forSubmission_20240128.csv", index=False )
```
